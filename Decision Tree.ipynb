# Decision Tree Algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
```

```python
iris = datasets.load_iris()
```

```python
x = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
x.head()
```

```python
x.info()
```

```python
x.describe()
```

```python
y = iris.target
y
```

```python
print (iris.target_names)
```

# Preparing the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

# Training the Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
print('Decision Tree Classifier Created Successfully')
```

# Predictions and Evaluation of Decision Tree

```python
predictions = dtree.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

# Visualizing the Decision Tree

```python
from sklearn import tree
fn = ['sepal length (cm)' , 'sepal width (cm)' , 'petal length (cm)' , 'petal width (cm)']
cn = ['setosa' , 'versicolor' , 'verginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (5,5), dpi = 400)
tree.plot_tree(dtree, feature_names = fn, class_names = cn, filled = True);
```
